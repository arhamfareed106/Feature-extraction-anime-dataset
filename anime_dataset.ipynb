```python
def extract_episodes(txt):
    """
    Extracts the episode count from a string by finding the text within the first set of parentheses.

    Args:
        txt (str): The input string, expected to contain the episode count in parentheses.

    Returns:
        str: The extracted episode count as a string, or None if no parentheses are found.
    """
    check = False
    data = ""
    for i in txt:
        if i == "(":
            check = True
            continue
        elif i == ")":
            check = False
            return data.strip()  # stop after first pair
        if check:
            data += i
    return data.strip() if data else None
```

```python
# Apply the extract_episodes function to the 'Title' column to create the 'Episode' column
df['Episode']=  df['Title'].apply(extract_episodes)
```

## Extract anime type

### Subtask:
Parse the 'Title' column to extract whether the anime is a TV series, Movie, OVA, etc. This can provide insights into how the anime type relates to other features.


**Reasoning**:
Define a function to extract the anime type from the 'Title' column and apply it to create a new 'Type' column in the DataFrame.



```python
def extract_type(title):
    if "TV" in title:
        return "TV"
    elif "Movie" in title:
        return "Movie"
    elif "OVA" in title:
        return "OVA"
    elif "ONA" in title:
        return "ONA"
    else:
        return "Unknown"

df['Type'] = df['Title'].apply(extract_type)
display(df.head())
```

## Extract start and end years

### Subtask:
Extract the start and end years from the 'Total Time' column. This allows for analysis of anime trends over time.


**Reasoning**:
Define a function to extract start and end years from the 'Total Time' column and apply it to the DataFrame to create new 'Start Year' and 'End Year' columns.



```python
from datetime import datetime

def extract_years(period):
    try:
        start_str, end_str = period.split(" - ")
        start_year = datetime.strptime(start_str, "%b %Y").year
        end_year = datetime.strptime(end_str, "%b %Y").year
        return start_year, end_year
    except:
        return None, None

df[['Start Year', 'End Year']] = df['Total Time'].apply(lambda x: pd.Series(extract_years(x)))
display(df.head())
```

## Calculate average episodes per year

### Subtask:
Using the extracted start/end years and the 'Episode' count, calculate the average number of episodes released per year for each anime.


**Reasoning**:
Define a function to calculate the average episodes per year and apply it to the DataFrame to create a new column.



```python
def calculate_avg_episodes_per_year(row):
    start_year = row['Start Year']
    end_year = row['End Year']
    episode_count = row['Episode']

    if start_year is None or end_year is None or episode_count is None:
        return None

    # Convert episode count to integer if it's not already
    try:
        episode_count = int(episode_count)
    except ValueError:
        return None

    duration = end_year - start_year + 1

    if duration <= 0:
        return episode_count  # Handle case where start and end year are the same or invalid

    return episode_count / duration

df['Avg Episodes Per Year'] = df.apply(calculate_avg_episodes_per_year, axis=1)
display(df.head())
```

## Categorize score into bins

### Subtask:
Group the 'Score' into categories (e.g., "High", "Medium", "Low") to facilitate categorical analysis and visualization.


**Reasoning**:
Define a function to categorize scores and apply it to the DataFrame to create a new 'Score Category' column.



```python
def categorize_score(score):
    if score >= 8.0:
        return "High"
    elif score >= 7.0 and score < 8.0:
        return "Medium"
    elif score >= 6.0 and score < 7.0:
        return "Average"
    else:
        return "Low"

df['Score Category'] = df['Score'].apply(categorize_score)
display(df.head())
```

## Analyze member count

### Subtask:
If available in the 'Title' column (as seen in some examples), extract and clean the member count to understand popularity.


**Reasoning**:
Define a function to extract the member count from the 'Title' column, apply the function to create the 'Member Count' column, convert the column to numeric, and display the head of the dataframe.



```python
import re

def extract_member_count(title):
    # Look for a pattern of numbers followed by " members" after a closing parenthesis
    match = re.search(r'\)\s*([\d,]+)\s*members', title)
    if match:
        # Remove commas from the matched number and return it
        return match.group(1).replace(',', '')
    return None

df['Member Count'] = df['Title'].apply(extract_member_count)
df['Member Count'] = pd.to_numeric(df['Member Count'], errors='coerce')
display(df.head())
```
